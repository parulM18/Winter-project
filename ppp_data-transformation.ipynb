```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df_I=pd.read_csv("ppp_India.csv")
df_I
```

```python
df_I= df_I.rename({'Value': 'ppp_India'}, axis=1)
del df_I['LOCATION']
```

```python
df_I
```

```python
df_C=pd.read_csv("ppp_China.csv")
df_C= df_C.rename({'Value': 'ppp_China'}, axis=1)
del df_C['LOCATION']
```

```python
df_US=pd.read_csv("ppp_USA.csv")
df_US= df_US.rename({'Value': 'ppp_USA'}, axis=1)
del df_US['LOCATION']
```

```python
df_UK=pd.read_csv("ppp_UK.csv")
df_UK= df_UK.rename({'Value': 'ppp_UK'}, axis=1)
del df_UK['LOCATION']
```

```python
df_J=pd.read_csv("ppp_Japan.csv")
df_J= df_J.rename({'Value': 'ppp_Japan'}, axis=1)
del df_J['LOCATION']
```

```python
from functools import reduce
df= [df_I, df_C, df_US, df_UK, df_J]
df = reduce(lambda left, right: pd.merge(left, right, on='TIME'), df)
```

```python
df
```

```python
df.to_csv("PPP_country-wise.csv", index=False)
```

```python
df.plot(x="TIME", y=["ppp_India", "ppp_China"])
plt.show()
```
