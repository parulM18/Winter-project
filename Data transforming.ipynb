```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df1=pd.read_csv("Import.csv")
df1
```

```python
df1=(df1.loc[df1['Partner Name'].isin(['China','United States', 'Japan', 'Bangladesh', 'United Kingdom'])])
```

```python
df1
```

```python
df1=df1.melt(id_vars=["Reporter Name", "Partner Name", "Trade Flow", "Product Group", "Indicator"], 
        var_name="Year", 
        value_name="Import")
df1
```

```python
df2=pd.read_csv("Export.csv")
df2
```

```python
df2=(df2.loc[df2['Partner Name'].isin(['China','United States', 'Japan', 'Bangladesh', 'United Kingdom'])])
df2
```

```python
df2=df2.melt(id_vars=["Reporter Name", "Partner Name", "Trade Flow", "Product Group", "Indicator"], 
        var_name="Year", 
        value_name="Export")
df2
```

```python
df= pd.merge(df1, df2, on=["Year", "Partner Name"])
df
```

```python
df["Trade deficit"]= (df["Export"]-df["Import"])
df
```

```python
del df['Trade Flow_x']
del df['Product Group_x']
del df['Product Group_y']
del df['Reporter Name_y']
del df['Indicator_x']
del df['Indicator_y']
```

```python
del df['Trade Flow_y']
df
```
